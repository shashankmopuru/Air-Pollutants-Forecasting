Handling the missing values

Convert all -200 to NaN

Replace all NaN values with mean of that specific column

```python
air_quality_data = air_quality_data.replace(to_replace=-200, value=np.NaN)
```

```python
# checking the number of missing values in the dataframe
air_quality_data.isnull().sum()
```

This shows the actual number of missing values

```python
air_quality_data.tail()
```

```python
# Select only numeric columns
numeric_columns = air_quality_data.select_dtypes(include=np.number).columns
air_quality_numeric = air_quality_data[numeric_columns]

# Calculate the mean of numeric columns
air_quality_numeric.mean()
```

```python
# replacing the missing values with mean values
# Select only numeric columns
numeric_columns = air_quality_data.select_dtypes(include=np.number).columns

# Fill missing values in numeric columns with their respective means
air_quality_data[numeric_columns] = air_quality_data[numeric_columns].fillna(air_quality_data[numeric_columns].mean())
```

**Forecasting with FB Prophet** **Algorithm** **bold text**

```python
# converting the date from DD/MM/YYYY to YYYY.MM.DD

date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y') # Specify the correct format
print(date_info)
```

```python
time_info = air_quality_data['Time']
```

```python
print(time_info)
```

```python
time_info = time_info.apply(lambda x:x.replace('.', ':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# combining 2 series to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis=1)
```

```python
date_time.head()
```

```python
date_time.shape
```

```python
# combining date and time

date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)
```

```python
date_time.head()
```

```python
date_time.info()
```

we have to convert ds from object data type to date time format

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y'] = air_quality_data['T']
```

```python
data.head()
```

```python
! pip install prophet
```

```python
from prophet import Prophet
```

```python
# training the prophet model

model = Prophet()
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=7000, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].iloc[10000:10003]
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```

```python
df_merge = pd.merge(data, forecast[['ds','yhat']],on='ds')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print ("Mean Absolute Error:", mean_absolute_error(df_merge.y, df_merge.yhat))
```
